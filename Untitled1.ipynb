```python
import random

question_bank = [
    {
        "question": "What is the capital of France?",
        "options": ["Berlin", "Madrid", "Paris", "Lisbon"],
        "answer": "Paris"
    },
    {
        "question": "What is 2 + 2?",
        "options": ["3", "4", "5", "6"],
        "answer": "4"
    },
    {
        "question": "Who wrote 'Hamlet'?",
        "options": ["Shakespeare", "Dickens", "Hemingway", "Austen"],
        "answer": "Shakespeare"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["Venus", "Mars", "Earth", "Jupiter"],
        "answer": "Mars"
    },
    {
        "question": "What is the smallest prime number?",
        "options": ["0", "1", "2", "3"],
        "answer": "2"
    },
    {
        "question": "What is the largest mammal?",
        "options": ["Elephant", "Whale", "Giraffe", "Shark"],
        "answer": "Whale"
    },
    {
        "question": "What is the boiling point of water?",
        "options": ["90°C", "100°C", "110°C", "120°C"],
        "answer": "100°C"
    },
    {
        "question": "What is the square root of 64?",
        "options": ["6", "8", "10", "16"],
        "answer": "8"
    },
    {
        "question": "What is the capital of Japan?",
        "options": ["Beijing", "Seoul", "Tokyo", "Bangkok"],
        "answer": "Tokyo"
    },
    {
        "question": "Who painted the Mona Lisa?",
        "options": ["Van Gogh", "Da Vinci", "Picasso", "Monet"],
        "answer": "Da Vinci"
    }
]

def ask_question(question_data):
    print(f"\n{question_data['question']}")
    for idx, option in enumerate(question_data['options'], 1):
        print(f"{idx}. {option}")
    answer = input("Your answer (1/2/3/4): ").strip()

    if question_data['options'][int(answer)-1] == question_data['answer']:
        return True
    else:
        return False

def play_quiz():
    questions_to_ask = random.sample(question_bank, 5)  # Choose 5 random questions from the question bank
    correct_answers = 0

    for question_data in questions_to_ask:
        result = ask_question(question_data)
        if result:
            print("Correct!")
            correct_answers += 1
        else:
            print(f"Incorrect! The correct answer is: {question_data['answer']}")

    print(f"\nYour final score is: {correct_answers} out of 5.")

if __name__ == "__main__":
    play_quiz()
```
